```python
hap_tiwi_count={}
for line in open("../../SG_hap_annot/tiwi/tiwi_summary_star.csv"):
    line=line.strip()
    if line.startswith("Gene"):
        continue
    cols=line.split("\t")
    if "./." in cols[0]:
        continue
    inf=cols[0].split("*")
    gene=inf[0]
    dip="*"+"*".join(inf[1:])
    hap=dip.split("/")
    if gene not in hap_tiwi_count.keys():
        hap_tiwi_count[gene]={}
    count=int(cols[1])
    for h in hap:
        if h not in hap_tiwi_count[gene].keys():
            hap_tiwi_count[gene][h]=count
        else:
            hap_tiwi_count[gene][h]=hap_tiwi_count[gene][h]+count
```

```python
hap_tiwi_freq={}
for g in hap_tiwi_count.keys():
    hap_tiwi_freq[g]={}
    for h in hap_tiwi_count[g].keys():
        hap_tiwi_freq[g][h]=hap_tiwi_count[g][h]/234
```
