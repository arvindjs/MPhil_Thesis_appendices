```python
kg_pop={}
for line in open("20130606_g1k.ped"):
    line=line.strip()
    cols=line.split("\t")
    if line.startswith("Family"):
        print(line)
        continue
    kg_pop[cols[1]]=cols[6]
```

```python
for line in open("tiwi_1kg_andaman.fam"):
    line=line.strip()
    cols=line.split()
    if not ((cols[1].startswith("TIWI") or cols[1].startswith("JAR")) or cols[1].startswith("ONG")):
        print("\t".join([cols[1],kg_pop[cols[1]]]))
    elif cols[1].startswith("TIWI"):
        print("\t".join([cols[1],"TIWI"]))
    elif cols[1].startswith("JAR") or cols[1].startswith("ONG"):
        print("\t".join([cols[1],"Andamanese"]))
```
