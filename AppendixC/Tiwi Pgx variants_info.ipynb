```python
for line in open("assoc_test/Tiwi_187_pgx1a1b.vcf"):
    line=line.strip()
    if line.startswith("#"):
        continue
    cols=line.split("\t")
    none=0
    het=0
    hom=0
    i=0
    for c in cols[9:]:
        if i==15 or i==27 or i==115:
            i=i+1
            continue
        geno=c.split(":")[0]
        if geno=="0/0":
            none=none+1
        elif geno=="0/1":
            het=het+1
        elif geno=="1/1":
            hom=hom+1
        i=i+1
    print("\t".join(cols[0:7]+[str(none),str(het),str(hom)]))
```
